# TUPLAS


```python
# As tuplas usam parênteses como sintaxe
tp = ("Banana", "Maçã", 10, 50)
```

```python
# Retornando o primeiro elemento
tp[0]
```

```python
#Diferentes das listas, tuplas são imutáveis, o que quer dizer que não podemos alterar os seus elementos
tp[0] = "Laranja"
```

```python
tp.count("Maçã")
```

```python
tp[0:2]
```

#Dicionários



```python
# Para criar um dicionário utlizamos as {}
dc = {"Maçã": 20, "Banana": 10, "Laranja": 15, "Uva": 5} #Dicionários trabalham com o conceito (chave e valor)
dc
```

```python
#Acessando o valor de um dicionário atráves da chave
dc["Maçã"]
```

```python
#Atualizando o valor de Maçã
dc["Maçã"] = 25
dc
```

```python
#Retornando todas as chaves do dicionário
dc.keys()
```

```python
#retornando os valores do dicionário
dc.values()
```

```python
#verificando se já existe um chave no dicionário e, caso não exista, ele irá inserir
dc.setdefault("Limão", 22)
```
