```python
#criando nosso dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#verificando os tipos de dados
df.dtypes
```

```python
#qual a receita total
df["Valor Venda"].sum()
```

```python
#Qual custo total
df["Custo"] = df["Custo Unit谩rio"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Qual o custo total arrendonda para receber apenas duas casa decimais
round(df["Custo"].sum(),2)
```

```python
#Agora que temos a receita e o custo total, podemos achar o lucro total
#Vamos criar uma coluna Lucro que ser谩 Receita - Custo
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
# Total lucro
round(df["Lucro"].sum(),2)
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

Agora, queremos saber a m茅dia do tempo de envio de cada Marca. E para isso, precisamos transformr a coluna Tempo_envio em n煤merica

```python
# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo de coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#Media do tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```
